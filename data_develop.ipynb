### test user_id

```python
listing_info_userid_idx = listing_info_data['user_id'].value_counts().index
train_userid_idx = train_data['user_id'].value_counts().index
test_userid_idx = test_data['user_id'].value_counts().index
user_info_userid_idx = user_info_data['user_id'].value_counts().index
user_taglist_userid_idx = user_taglist_data['user_id'].value_counts().index
user_repay_logs_userid_idx = user_repay_logs_data['user_id'].value_counts().index
user_behavior_logs_userid_idx = user_behavior_logs_data['user_id'].value_counts().index
```

```python
def find_diff_index(idx1, idx2):
    return idx1.difference(idx2)
```

```python
print('||listing_info    :||',find_diff_index(train_userid_idx, listing_info_userid_idx), '\n')
print('||user_taglist    :||',find_diff_index(train_userid_idx, user_taglist_userid_idx), '\n')
print('||user_behavior_logs :||',find_diff_index(train_userid_idx, user_behavior_logs_userid_idx), '\n')
print('||user_repay_logs :||',find_diff_index(train_userid_idx, user_repay_logs_userid_idx), '\n')
print('||user_info       :||',find_diff_index(train_userid_idx, user_info_userid_idx), '\n')
```

```python
print('||listing_info    :||',find_diff_index(test_userid_idx, listing_info_userid_idx), '\n')
print('||user_taglist    :||',find_diff_index(test_userid_idx, user_taglist_userid_idx), '\n')
print('||user_behavior_logs :||',find_diff_index(test_userid_idx, user_behavior_logs_userid_idx), '\n')
print('||user_repay_logs :||',find_diff_index(test_userid_idx, user_repay_logs_userid_idx), '\n')
print('||user_info       :||',find_diff_index(test_userid_idx, user_info_userid_idx), '\n')
```
